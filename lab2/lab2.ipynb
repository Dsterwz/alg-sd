Реализовать следующие структуры данных:
  
● <b> Стек (stack): </b>  
Операции для стека: 
<i> инициализация, проверка на пустоту, добавление нового элемента в начало, извлечение элемента из начала;  </i>  
  
● <b> Дек (двусторонняя очередь, deque):  </b>  
Операции для дека: 
<i> инициализация, проверка на пустоту, добавление нового элемента в начало, добавление нового элемента в конец, извлечение элемента из начала, извлечение элемента из конца. </i>  
  
<i> Разработать программу обработки данных, содержащихся в заранее подготовленном txt-файле, в соответствии с заданиями, применив указанную в задании структуру данных. Результат работы программы вывести на экран и сохранить в отдельном txt-файле. </i>

```python
class Stack:
    
    def __init__(self, s):
        self.s = s
      
    def is_empty(self) -> bool:
        return len(self.s) == 0
        
    def add(self, value):
        self.s.append(value)

    def extract(self):
        return self.s.pop()

class Deque:

    def __init__(self, d):
        self.d = d

    def is_empty(self) -> bool:
        return len(self.d) == 0

    def add_end(self, value):
        self.d.append(value)

    def add_start(self, value):
        self.d.insert(0, value)

    def extract_end(self):
        return self.d.pop()
    
    def extract_start(self):
        return self.d.pop(0)
```

```python
first_deque = Deque([])
second_deque = Deque([])


def cocktail(other, another):
    min_name = chr(126)

    while not another.is_empty():
        name = another.extract_start()
        for i in range(len(min_name)):
            if name[i] < min_name[i]:
                other.add_start(name)
                min_name = name
                break
            elif name[i] > min_name[i]:
                other.add_end(name)
                break

    return other.extract_start()


with open('books.txt', 'r') as f:
    for line in f:
        second_deque.add_end(line.strip())
    drug = True
    while not second_deque.is_empty() or not first_deque.is_empty():
        if drug:
            print(cocktail(first_deque, second_deque))
            drug = False
        else:
            print(cocktail(second_deque, first_deque))
            drug = True
```

### Задание №3
Даны три стержня и n дисков различного размера. Диски можно надевать на стержни, образуя из них башни. Перенести n дисков со стержня А на стержень С, сохранив их первоначальный порядок. При переносе дисков необходимо соблюдать следующие правила:
- на каждом шаге со стержня на стержень переносить только один диск;
- диск нельзя помещать на диск меньшего размера;
- для промежуточного хранения можно использовать стержень В.
Реализовать алгоритм, используя три стека вместо стержней А, В, С. Информация о дисках хранится в исходном файле.

```python
from google.colab import files
files.upload()
```

```python
a, b, c = Stack([]), Stack([]), Stack([])


def hanoi(n, start, temp, end):
    if n <= 1: return
    elif n == 2:
        temp.add(start.extract())
        end.add(start.extract())
        end.add(temp.extract())
    else: 
        hanoi(n-1, start, end, temp)
        end.add(start.extract())
        hanoi(n-1, temp, start, end)


with open("towers_of_hanoi.txt", 'r') as f:
    for line in f:

        word = line.strip().split(" ")[::-1]
        n = len(word)

        for i in word:
            a.add(int(i))

        hanoi(n, a, b, c)
        

        while not c.is_empty():
            print("c: ", c.extract())

        while not a.is_empty():
            print("a: ", a.extract())

        while not b.is_empty():
            print("b: ", b.extract())
```

### Задание №6
Дан файл из символов. Используя стек, за один просмотр файла напечатать сначала все цифры, затем все буквы, и, наконец, все остальные символы, сохраняя исходный порядок в каждой группе символов.

```python
from google.colab import files
files.upload()
```

```python
s = Stack([])

with open("symbols.txt", 'r') as f:
    for line in f:
        word = len(line.strip())
        for i in range(word):
            s.add(line.strip()[-1*i-1])

    nums = ""
    alphas = ""
    alnums = ""

    while not s.is_empty():
        t = s.extract()

        if t.isdigit():
            nums += t

        if t.isalpha():
            alphas += t

        if not t.isalnum():
            alnums += t

        if not s.is_empty():
            s.add(s.extract())
            
    print(nums + alphas + alnums)
```

```python
d = Deque([])

with open("nums.txt", 'r') as f:
    for line in f:
        word = line.strip().split(" ")[::-1]
        j = 0
        for i in word:
            if int(i) < 0:
                d.add_end(int(i))
                j += 1
            else:
                d.add_start(int(i))
                
    a = 0
    while not d.is_empty():
        if a < j:
            print(d.extract_end())
            a += 1
        else:
            print(d.extract_start())
```
